Создайте функцию, которая принимает как аргументы целое число N и первый элемент (вещественное число), и разность (вещественное число) и создает матрицу numpy по диагонали, которой располагаются первые N членов арифметической прогрессии.

```python
def create_diagonal_matrix(n, first_element,d):
    diag=[first_element+d*i for i in range(n)]
    matrix = np.diag(diag)
    return matrix
d = create_diagonal_matrix(7, 2, 4)
d
```

Создайте двумерный массив, содержащий единицы на границе и нули внутри.

```python
def create_border_array(size):
    array = np.zeros((size, size),dtype='int')
    array[0, :] = 1
    array[-1, :] = 1
    array[:, 0] = 1
    array[:, -1] = 1
    
    return array

A= create_border_array(5)
A
```

Создайте две матрицы размером (5,5). Одна матрица содержит 5 в шахматном порядке как в задаче домашнего задания, другая имеет треугольную форму содержащую 5 на основной диагонали и в позициях выше ее, а ниже все 0. Посчитайте их детерминант и найдите обратные матрицы.

```python
chess_matrix = np.zeros((5, 5))
chess_matrix[::2, ::2] = 5
chess_matrix[1::2, 1::2] = 5
triangle_matrix = np.tri(5, 5, k=0) * 5
det_chess = np.linalg.det(chess_matrix)
det_triangle = np.linalg.det(triangle_matrix)
inv_chess_matrix = np.linalg.inv(chess_matrix) if det_chess != 0 else "Невозможно вычислить"
inv_triangle_matrix = np.linalg.inv(triangle_matrix) if det_triangle != 0 else "Невозможно вычислить"
chess_matrix, triangle_matrix, det_chess, det_triangle, inv_chess_matrix, inv_triangle_matrix
```

Выделите в отдельную таблицу последние десять строк. Уберите в ней столбец с ценой. Склейте ее с первоначальной таблицей при помощи append. Заполните отсутствующие значения цены средним по таблице.

```python
last_ten_rows = data_renamed.tail(10)
last_ten_rows_wp = last_ten_rows.drop(columns=['Цена'])
combined_data = data_renamed.append(last_ten_rows_wp, ignore_index=True)
average_price = combined_data['Цена'].mean()
combined_data['Цена'].fillna(average_price, inplace=True)
combined_data
```

Выделите пять последних колонок в отдельную таблицу. Удалите в ней строки, в которых цена ниже среднего. Присоедините эту таблицу к изначальной (выберите самый подходящий тип соединения).

```python
last_five_columns = combined_data.iloc[:, -5:]
average_price = combined_data['Цена'].mean()
filtered_last_five_columns = last_five_columns[last_five_columns['Цена'] >= average_price]
merged_data = combined_data.merge(filtered_last_five_columns, on=combined_data.columns[-5:].tolist(), how='left')
merged_data.tail()
```

Выведите таблицу, содержащую среднюю цену и количество квартир на каждом этаже из первоначального набора данных
